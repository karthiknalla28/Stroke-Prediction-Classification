```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Importing and reading the dataset as df 
df = pd.read_csv('/content/Stroke dataset.csv')
df
```

**Data Cleaning**

```python
# Caluclating the sum of missing values 
df.isnull().sum()
```

**We have learnt that the 'bmi' attribute has 201 missing values.**<br>
We need to resolve this issue before we go any further.

```python
# Dropping the null values in 'BMI' coulmn
df.dropna(subset=['bmi',],inplace = True)
df
```

**Outliers detection and correction**

```python
# function to detect outliers
def Outliers_Dtetector(attribute, attribute_name):
   Q1 = np.percentile(attribute, 25, interpolation = 'midpoint') 
   Q3 = np.percentile(attribute, 75, interpolation = 'midpoint') 
   IQR = Q3 - Q1
   lower_fourth = Q1 - (1.5 * IQR)
   upper_fourth = Q3 + (1.5 * IQR)
   outliers = attribute[((attribute < lower_fourth) | (attribute > upper_fourth))]
   
   print(f'Name:{attribute_name}')
   print(f'Number of Outliers: {len(outliers)}')
   print(f'Statistical Lower fourth: {lower_fourth:.3f}')
   print(f'Statistical Upper fourth: {upper_fourth:.3f}')

Outliers_Dtetector(df['bmi'], 'bmi')
print("")
Outliers_Dtetector(df['avg_glucose_level'], 'avg_glucose_level')
```

```python
# variables to store median values of bmi and 
median_bmi = np.nanmedian(df['bmi'])
median_avg_glucose_level = np.nanmedian(df['avg_glucose_level'])
```

```python
# Replacing outliers with median values
df['bmi'].iloc[(df['bmi'] > 46.29) | (df['bmi'] < 10.30)] = median_bmi
df['avg_glucose_level'].iloc[(df['avg_glucose_level'] > 169.35) | (df['avg_glucose_level'] < 21.98)] = median_avg_glucose_level
```

```python
#dropping the column id
df = df.drop('id', axis=1)
df
```

**Label Encoding**

```python
# Import preprocessing from sklearn for Label Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()


# Encode labels in column 'gender'.
# Male : 1 , Female : 0
df['gender']= label_encoder.fit_transform(df['gender'])


# Encode labels in column 'ever_married'.
# Yes : 1 , No : 0
df['ever_married']= label_encoder.fit_transform(df['ever_married'])


# Encode labels in column 'Residence_type'.
# Urban : 1, Rural : 0
df['Residence_type']= label_encoder.fit_transform(df['Residence_type'])


# Encode labels in column 'work_type'.
# Govt_job : 0, Never_worked : 1,Private : 2,Self-employed : 3  children : 4, 
df['work_type']= label_encoder.fit_transform(df['work_type'])


# Encode labels in column 'smoking_status'.
# Unknown : 0, formerly smoked : 1, never smoked : 2, smokes : 3
df['smoking_status']= label_encoder.fit_transform(df['smoking_status'])


df
```
