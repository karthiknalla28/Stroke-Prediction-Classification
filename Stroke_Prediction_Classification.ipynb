**Data Preprocessing**

**Data Cleaning**

```python
# Caluclating the sum of missing values 
df.isnull().sum()
```

```python
# Dropping the null values in 'BMI' coulmn
df.dropna(subset=['bmi'],inplace = True)
df
```

```python
# Making sure that dataset has no missing data
df.isnull().sum()
```

**Label Encoding**

```python
# Import preprocessing from sklearn for Label Encoding
from sklearn import preprocessing
label_encoder = preprocessing.LabelEncoder()


# Encode labels in column 'gender'.
# Male : 1 , Female : 0
df['gender']= label_encoder.fit_transform(df['gender'])


# Encode labels in column 'ever_married'.
# Yes : 1 , No : 0
df['ever_married']= label_encoder.fit_transform(df['ever_married'])


# Encode labels in column 'Residence_type'.
# Urban : 1, Rural : 0
df['Residence_type']= label_encoder.fit_transform(df['Residence_type'])


# Encode labels in column 'work_type'.
# Govt_job : 0, Never_worked : 1,Private : 2,Self-employed : 3  children : 4, 
df['work_type']= label_encoder.fit_transform(df['work_type'])


# Encode labels in column 'smoking_status'.
# Unknown : 0, formerly smoked : 1, never smoked : 2, smokes : 3
df['smoking_status']= label_encoder.fit_transform(df['smoking_status'])


df
```

```python
def plot_hist(col, bins=30, title="",xlabel="",ax=None):
    sns.distplot(col, bins=bins,ax=ax)
    ax.set_title(f'Histogram of {title}',fontsize=20)
    ax.set_xlabel(xlabel)
```

```python
fig, axes = plt.subplots(1,2,figsize=(11,7),constrained_layout=True)
plot_hist(df.bmi,
          title='Bmi',
          xlabel="Level of the BMI",
          ax=axes[0])
plot_hist(df.age,
          title='Age',
          xlabel='Age',
          ax=axes[1])
plt.show()
```


Observation

Both the private sector and the self-employed have an equal number of stroke cases. However, as compared to the first two categories, those who work in government are more likely to avoid having a stroke. Children are also less likely to suffer from a stroke. Perhaps the level of pressure experienced by the workers can explain that.

Surprisingly, it appears that smoking does not significantly increase the risk of stroke, as the proportion of stroke victims is almost the same regardless of smoking status.



```python
plt.figure(figsize=(17,7))
sns.catplot(x="gender", y="stroke", hue="heart_disease", kind="bar", data=df)
sns.catplot(x="gender", y="stroke", hue="Residence_type", kind="bar", data=df)
sns.catplot(x="gender", y="stroke", hue="hypertension",  kind="bar", data=df)
plt.show()
```

Observation .

The plots demonstrate that a person's gender is not a factor that determines whether they will experience a stroke or not.

Hypertension and heart disease are correlated to stroke 

It appears that pollution is connected with having a stroke since rural individuals are less likely to suffer from strokes than urban ones.
